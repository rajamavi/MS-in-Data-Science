```python
import pandas as pd
import matplotlib.pyplot as plt

# Load generated data
user_profiles = pd.read_csv(r'C:\Users\Olivia\Documents\Fall-2023\Big-Data-Technology\Big-Daa_Project\UserProfiles_data.csv')
product_data = pd.read_csv(r'C:\Users\Olivia\Documents\Fall-2023\Big-Data-Technology\Big-Daa_Project\ProductData_data.csv')
user_interactions = pd.read_csv(r'C:\Users\Olivia\Documents\Fall-2023\Big-Data-Technology\Big-Daa_Project\UserInteractions_data.csv')

# User Profiles Report
def user_profiles_report():
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 3, 1)
    user_profiles['Age'].plot(kind='hist', bins=20, title='Age Distribution')

    plt.subplot(1, 3, 2)
    user_profiles['Gender'].value_counts().plot(kind='bar', title='Gender Distribution')

    plt.subplot(1, 3, 3)
    user_profiles['Income'].plot(kind='hist', bins=20, title='Income Distribution')

    plt.tight_layout()
    plt.show()

# Product Data Report
def product_data_report():
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    product_data['Category'].value_counts().plot(kind='bar', title='Category-wise Product Count')

    plt.subplot(1, 2, 2)
    product_data['Price'].plot(kind='hist', bins=20, title='Price Distribution')

    plt.tight_layout()
    plt.show()

# User Interactions Report
def user_interactions_report():
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 3, 1)
    user_interactions['Product_ID'].value_counts().head(10).plot(kind='bar', title='Top 10 Popular Products')

    plt.subplot(1, 3, 2)
    user_interactions['Interaction_Type'].value_counts().plot(kind='bar', title='Interaction Type Distribution')

    plt.subplot(1, 3, 3)
    user_interactions['Timestamp'] = pd.to_datetime(user_interactions['Timestamp'])
    user_interactions.set_index('Timestamp')['User_ID'].resample('D').count().plot(kind='line', title='Daily Interaction Trend')

    plt.tight_layout()
    plt.show()

# Generate and display reports
user_profiles_report()
product_data_report()
user_interactions_report()
```
